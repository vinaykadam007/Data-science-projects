```python
import pandas as pd
from pandas import DataFrame, read_csv
from matplotlib import pyplot as plt
import matplotlib, sys
```

```python
print('Python version ' + sys.version)
print('Pandas version ' + pd.__version__)
print('Matplotlib version ' + matplotlib.__version__)
```

```python
#initial data set 
names = ['Bob','Jessica','Mary','John','Mel']
births = [968, 155, 77, 578, 973]
BabyDataSet = list(zip(names, births))
```

```python
print(BabyDataSet)
```

```python
#create a pandas data frame now. it is similar to an excel sheet with columns
df = DataFrame(data=BabyDataSet, columns=['Names', 'Births'])
df
```

```python
#save the dataset to a csv and read from the same csv file, will be saved with headers and index
#same for excel to_excel function
df.to_csv("birthsdataSet.csv", index=False)
```

```python
#read the csv again 
#if headers are not there, then header=None
#if headers are there in file, need not mention anything
df = pd.read_csv("birthsdataSet.csv")
print("read data from csv\n", df)
```

```python
print(df.dtypes)
```

```python
print(df.Births.dtype)
```

```python
#sorting of dataframes, some basic analysis
SortedBabyDataSet = df.sort_values(['Births'])
print(SortedBabyDataSet)
print("**"*15)
print(SortedBabyDataSet['Births'][2:])
print(SortedBabyDataSet['Births'].mean())
```

```python
#sort by name first
NameSortedBabyDataSet = df.sort_values(['Names'])
print(NameSortedBabyDataSet)
print(NameSortedBabyDataSet.head(1))
```

```python
print("max value of birth", df['Births'].max(), "\n")

#entire list of information
print("Names in data frame\n", df['Names'], "\n")
print("Births in data frame\n", df['Births'])
```
