```python
max_k = 6
for k in range(2, max_k):
    print("Running Kmeans with {0} clusters".format(k))
    kmeans = KMeans(n_clusters = k, init = 'k-means++', max_iter = 300, n_init = 10, n_jobs=2)
    y_kmeans = kmeans.fit_predict(df.iloc[:, 1:])
    print("Clusters found in {0} iterations".format(kmeans.n_iter_))
    
    for i in range(k):
        print("Items in Cluster {0}: {1}".format(i+1, len([j for j in y_kmeans==i if j]) ))
```
