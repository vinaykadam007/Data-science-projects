```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("./Indian_Student_Data.csv")
df.head()
```

```python
# --- Build Map ---
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt
import numpy as np

def get_marker_color(magnitude):
    if magnitude < 10.0:
        return ('go')
    elif magnitude < 20.0:
        return ('yo')
    else:
        return ('ro')

# Make this plot larger.
plt.figure(figsize=(16,12))

eq_map = Basemap(projection='robin', resolution = 'l', area_thresh = 1000.0,
             lat_0=0, lon_0=0)
eq_map.drawcoastlines()
eq_map.drawcountries()
eq_map.fillcontinents(color = 'whitesmoke')
eq_map.drawmapboundary()
eq_map.drawmeridians(np.arange(0, 360, 30))
eq_map.drawparallels(np.arange(-90, 90, 30))

# min_marker_size = 2.5 
# for lon, lat, mag in zip(lons, lats, magnitudes):
for sr, row in df[["latitude","longitude","percentage"]].iterrows():
    x,y = eq_map(row.longitude, row.latitude)
    msize = int((row.percentage/20) + 5.0)
    marker_string = get_marker_color(row.percentage)
    eq_map.plot(x, y, marker_string, markersize=msize)
   
title_string = "Indian Student Distribution\n"
plt.title(title_string)

plt.show()
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_samples, silhouette_score
```

```python
kmeans = KMeans(n_clusters=7, init='k-means++')
y_kmeans = kmeans.fit_predict(df[['latitude', 'longitude', 'percentage']])
y_kmeans
```

```python
print("Clusters found in {0} iterations".format(kmeans.n_iter_))
```

```python
silhouette_avg = silhouette_score(df[['latitude', 'longitude', 'percentage']], y_kmeans)
print(silhouette_avg)
```

```python
#Finding the optimum number of clusters for k-means classification
wcss = []

#wcss = within cluster sum of squares 

max_clusters = 10
for i in range(2, max_clusters):
    kmeans = KMeans(n_clusters = i, init = 'k-means++')
    kmeans.fit(df[['latitude', 'longitude', 'percentage']])
    wcss.append(kmeans.inertia_)
    
#Plotting the results onto a line graph, allowing us to observe 'The elbow'
plt.plot(range(2, max_clusters), wcss)
plt.title('The elbow method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS') #within cluster sum of squares
plt.grid()
plt.xticks(range(2, max_clusters), range(2, max_clusters))
plt.show()
```
