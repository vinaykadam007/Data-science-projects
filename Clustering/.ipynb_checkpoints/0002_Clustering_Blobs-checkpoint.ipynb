```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()  # for plot styling
import numpy as np
```

## Making dataset

```python
from sklearn.datasets.samples_generator import make_blobs
X, y_true = make_blobs(n_samples=300, centers=4,
                       cluster_std=0.9, random_state=0)
# print(X)
plt.scatter(X[:, 0], X[:, 1], s=50);
```

## Running Kmeans using scikit learn 

## nclusters = Number of clusters to form
## max_iter = Maximum iterations for a single run
## n_init = Number of times results will be run for a new centroid
## init = Method for clustering

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=4, max_iter=300, n_init=5, init='random')
kmeans.fit(X)
y_kmeans = kmeans.predict(X)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='viridis')

centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=200);
```
