```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import LinearRegression
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
df = pd.read_csv('./data/auto-mpg.csv')
(df.head())
```

```python
cols = list(df.columns)
print(cols)
```

```python
df = df[df['horsepower'] != '?']
df['horsepower'] = df['horsepower'].apply(lambda x: float(x))
```

```python
import seaborn as sb

plt.figure(figsize=(10, 8))
sb.heatmap(df.corr(), annot=True, fmt='.2f')
```

```python
for col in cols[1:]:
    plt.figure()
    plt.scatter(df[col], df['mpg'])
    plt.xlabel(col)
    plt.ylabel("MPG")
    plt.grid()
```

```python
# applying linear regression on some features
X = df[['cylinders', 'displacement', 'horsepower', 'weight']]
y = df['mpg']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2340)

lm = LinearRegression()
lm.fit(X_train, y_train)
print("Score is", lm.score(X, y))
print("MSE is", mean_squared_error(y_test, lm.predict(X_test)))
```

# Single variable Regression Bonus

```python
X = df['weight']
y = df['mpg']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2340)
lm = LinearRegression()
lm.fit(X_train.values.reshape(-1, 1), y_train)
print(lm.score(X.values.reshape(-1, 1), y))
print("MSE is", mean_squared_error(y_test, lm.predict(X_test.values.reshape(-1, 1))))

plt.scatter(X_train, y_train)
plt.grid()
plt.plot([1500, 5010], lm.predict([[1500], [5010]]), c='r', linewidth=3)
plt.xlabel("Weight")
plt.ylabel("MPG")
```
