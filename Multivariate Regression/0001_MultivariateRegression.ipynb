## Multivariate Linear Regression
=======

Simple linear regression can be used to model a linear relationship between one response variable and one explanatory variable

Predicting Pizza prices

Here there are two explanaotory variable, pizza size and toppings


```python
%matplotlib inline
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

X = [3, 5, 6, 10, 14]
Y = [5, 5, 2, 2, 1]
Z = [5, 8, 12, 18, 22]

fig0 = plt.figure()
ax = fig0.add_subplot(111, projection='3d')
ax.scatter(X, Y, Z, c='skyblue', s=60)
ax.view_init(30, 30)
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
import pandas as pd
import numpy as np

df = pd.DataFrame({"X":X,"Y":Y,"Z":Z}, columns=["X", "Y", "Z"])
df.head()

# # Create and fit the model
model = LinearRegression()
model.fit(df[["X", "Y"]], df["Z"] )

fig0 = plt.figure()
xx = np.linspace(0, 8, 9)
yy = np.linspace(0, 20, 21)
mesh = np.meshgrid(xx, yy)
np.c_[xx.ravel(), yy.ravel()]
Z = model.predict(np.c_[xx.ravel(), yy.ravel()]).reshape(xx.shape)

print(model.predict(mesh))
ax = fig0.add_subplot(111, projection='3d')
p = ax.plot_surface(X, Y, Z, rstride=4, cstride=4)
```
