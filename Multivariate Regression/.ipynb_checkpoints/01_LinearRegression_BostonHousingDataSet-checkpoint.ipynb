```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
#read csv
cols = ['CRIM', "ZN", "INDUS", "CHAS", "NOX", "RM" , "AGE", "DIS", "RAD", "TAX", "PTRATIO", "B", "LSTAT", "MEDV"]
df = pd.read_csv('./data/housing.csv', names=cols, header=None)
print(df.head())
print(df.shape)
```

```python
df.isnull().any()
```

```python
df = df.dropna()
df.shape
```

```python
# df.corr()
sb.heatmap(df.corr(), annot=True, fmt='.1f')
```

```python
plt.figure()
plt.plot(df['RM'], df['MEDV'], linestyle='None', marker='o')
plt.xlabel('Average room per dwelling')
plt.ylabel('Median price of home')

plt.figure()
plt.plot(df['LSTAT'], df['MEDV'], linestyle='None', marker='o', color='r')
plt.xlabel('% lower status of population')
plt.ylabel('Median price of home')

plt.figure()
plt.plot(df['CRIM'], df['MEDV'], linestyle='None', marker='o', color='g')
plt.xlabel('Crime rate')
plt.ylabel('Median price of home')

plt.figure()
plt.plot(df['PTRATIO'], df['MEDV'], linestyle='None', marker='o', color='orange')
plt.xlabel('Pupil Teacher ratio')
plt.ylabel('Median price of home')

plt.show()
```

### Implementing Linear Regression

```python
lm = LinearRegression()

#Important functions
#lm.fit()
#lm.predict()
#lm.score()

#data set distribution
X = df[cols].drop('MEDV', axis=1)
y = df['MEDV']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=1)
```

```python
#get the fitting curve
lm_model = lm.fit(X_train, y_train)
```

```python
obtained_y_test = lm.predict(X_test)
```

```python
# plt.plot(X_test['LSTAT'], y_test, linestyle='None', marker='x', color='r')
pointa = X.min()
pointb = X.max()

temp =  lm.predict([pointa, pointb])
plt.plot([pointa, pointb],temp)
plt.xlabel("X Test")
plt.ylabel("Y Test")
plt.show()

#scatter plot of obtained_y_test with y_test
plt.scatter(y_test, obtained_y_test)
plt.xlabel("Test Prices")
plt.ylabel("Predicted Prices")
plt.show()
```

```python
#mean square error
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y_true = y_test, y_pred = obtained_y_test)
print("MSE = %f" % mse)
```

```python
lm.score(X_test, y_test)
```

```python
#the more folds we have, we will be reducing the error due the bias but increasing the error due to variance
from sklearn.model_selection import cross_val_score, cross_val_predict
from sklearn import metrics

scores = cross_val_score(lm, X, y, cv=5)
print("Cross-validated scores:", scores)
print("Accuracy: %0.2f (+/- %0.2f)" % (scores.mean(), scores.std() * 2))
```

```python
from sklearn import metrics
predictions = cross_val_predict(lm, X, y, cv=5)
plt.scatter(y, predictions)
```
